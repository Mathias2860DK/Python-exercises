```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(
    'owid-covid-data.csv')
```

```python
df
```

**Hvilket land i europa har testet flest personer pr. tusinde indbyggere (total_tests_per_thousand), siden juli 2021 til nu?**

```python
mask = df['continent'] == 'Europe'

max_cases = df[mask].groupby('location').agg({'total_tests_per_thousand': [
    'max']})
print(max_cases.sort_values([('total_tests_per_thousand', 'max')], ascending=False))
```

**Vis en graph af de ti lande med flest døde til dags dato.**

```python
#df[df['continent'].notna()].sort_values('total_cases', ascending=False).head(10)
top_10 = df[df['continent'].notna()].groupby('location').agg({'total_deaths': ['max']}).sort_values([('total_deaths', 'max')], ascending=False).head(10)

top_10.columns = ['total_deaths']
print(top_10.columns)
print(type(top_10))
top_10



```
