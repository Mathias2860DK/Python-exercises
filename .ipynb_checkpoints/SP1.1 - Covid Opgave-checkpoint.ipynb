#### 2. Hvilket land i europa har test flest personer pr. tusinde indbyggere (total_tests_per_thousand), siden juli 2021 til nu?

```python
mask = df['continent'] == 'Europe'
max_cases = df[mask].groupby('location').agg({'total_tests_per_thousand': [
    'max']})
max_cases.sort_values([('total_tests_per_thousand', 'max')], ascending=False).head(10)
```

#### 3. Vis en graph af de ti lande med flest døde til dags dato.

```python
#top_10 = df[df['continent'].notna()].groupby('location').agg({'total_deaths': [
#    'max']}).sort_values([('total_deaths', 'max')], ascending=False).head(10)
#print(type(top_10))
df_copy = df.copy()
new_top_10 = df_copy[df_copy['continent'].notna()].groupby('location')['total_deaths'].max().sort_values(ascending=False)
df_copy = df.set_index('location')
df['total_deaths'].update(new_top_10)
new_top_10 = new_top_10.reset_index().head(10)
new_top_10

#horaizontal barchart 
#Flip dataframe dff=df[::-1]
fig, ax = plt.subplots(figsize=(15, 8))
ax.barh(new_top_10['location'], new_top_10['total_deaths'])
plt.xlabel('Number of Total Deaths in Millions')
plt.ylabel('Country')
```

#### 4. Vælg tre lande og vis udviklingen af døde i forhold til total antal smittede i procent fra d. 1 april 2020 til dags dato.

```python
#Denmark, Brazil and China
#Total deaths cases in the world
data2 = df.copy()
data2.date = pd.to_datetime(data2['date'])
data2 = data2.groupby('date').sum()
data2['new_deaths'].plot(figsize = (11, 5))
plt.title('Timeline new deaths in world')
plt.xlabel('Date')
plt.ylabel('New deaths')
```

```python
data3 = df.copy()
type(data3)

mask = data3['location'] == 'Brazil'
df_brazil = data3[mask]
df_brazil.date = pd.to_datetime(df_brazil['date'])
df_brazil = df_brazil.groupby('date').sum()

df_brazil['date'] = df_brazil.index
df_brazil.set_index('date', inplace=True)

plt.figure(figsize=(15, 5))
plt.plot(df_brazil.index, df_brazil['new_deaths'], "-b", label="BRA Deaths")
plt.plot(df_brazil.index, df_brazil['new_cases'], "-y", label="BRA Cases")
plt.legend(loc="upper left")
plt.title("Deaths/cases ",fontsize=14)
plt.ylabel('New cases/deaths ')
plt.xlabel('Date')
plt.show()
```
